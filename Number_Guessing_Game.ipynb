```python
import random

guess=random.randint(1,20)

print ('='*50)
print('\t\tNumber Guessing Game')
print('='*50)
print('\t\tGet ready to play!')
print('='*50)
print('Note: You only get three tries to win.\n')

for i in range (0,3):
    num=int(input('Enter your guess: '))

    if num==guess:
        print('='*50)
        print(f'You guessed it right, the number was {guess}!')
        print('You Win!!!')
        print('='*50)
        break
    elif num<guess:
        print('You guessed a low number.')
        print('Try Again!')
    elif num>guess:
        print('You guessed a high number.')
        print('Try Again!')
else:
    print('='*50)
    print('You Lose!')
    print(f'The correct number was: {guess}')

print('='*50)
print('\t\t    THE END')
print('='*50)
```
