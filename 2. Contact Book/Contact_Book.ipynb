```python
print('='*50)
print('\t\t  Contact Book')
print('='*50)

book={}

full_name=input('Enter your name: ')
ph_no=input('Enter your phone number: ')
book.update({full_name:ph_no})
print('='*50)

print('\t\tContact Book Functions')
print('\n1. Add a contact')
print('2. Search a contact')
print('3. Update a contact')
print('4. Delete a contact')
print('5. View the contact book')
print('6. Exit')

serial=int(input('Enter the fucntion number: '))
print('='*50)

def main():
    global serial
    print('\t\tContact Book Functions')
    print('\n1. Add a contact')
    print('2. Search a contact')
    print('3. Update a contact')
    print('4. Delete a contact')
    print('5. View the contact book')
    print('6. Exit')
    serial=int(input('Enter the fucntion number: '))
    print('='*50)
    
def add():
    print('Enter the Contact Details to store')
    print('='*50)
    name=input('Enter the Name: ')
    number=input('Enter the Contact Number: ')
    book.update({name:number})
    print(f'The {name}:{number} has been added successfully')
    print('='*50)
    main()
    return

def search():
    sname=input('Enter the name you want to search: ')
    print(f"The Contact Number is: {book.get(sname, 'Not Found!')}")
    print('Search completed successfully!')
    print('='*50)
    main()
    return

def update():
    print('Enter the contact details to update')
    print('1. Name')
    print('2. Contact Number')
    choice=int(input('What you need to update(Enter the option no.):'))
    print('='*50)

    if choice ==1:
        pname=input('Enter the old name: ')
        book.pop(pname)
        uname=input('Enter the correct Name: ')
        unum=input('Enter the Contact Number: ')
        book.update({uname:unum})
        print('The name has been updated successfully!')
        print('='*50)
    else:
        uname=input('Enter the Name: ')
        unum=input('Enter the correct Contact Number: ')
        book.update({uname:unum})  
        print('The contact number has been updated successfully!')
        print('='*50)
    main()
    return
    
def delete():
    dname=input('Enter the name you want to delete: ')
    book.pop(dname)
    print(f'{dname} contact details deleted successfully!')
    print('='*50)
    main()
    return

def view_book():
    print('Contact Details in your Book:')
    print('='*50)
    if not book:
        print("Your contact book is empty!")
    else:
        for name, number in book.items():
            print(f"Contact: {name} | Phone: {number}") # REMOVED the break statement
    print('='*50)
    main()

def exit_book():
    print('You have successfully exited the contact book')
    print('Visit Again Later!')
    print('='*50)
    
while True:
    if serial==1:
        add()
    elif serial==2:
        search()
    elif serial==3:
        update()
    elif serial==4:
        delete()
    elif serial==5:
        view_book()
    elif serial==6:
        exit_book()
        break
    else:
        print('You have selected an Invalid Option.')
        dec= input('Do you want to visit Main Menu (Y/N): ')
        if dec.upper()=='Y':
            main()
        else:
            print('='*50)
            exit_book()
            break
```
